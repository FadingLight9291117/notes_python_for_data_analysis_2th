### Figure和Subplot

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--') # 黑色虚线
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig, axes = plt.subplots(2, 3)
```

```python
axes
```

```python
axes[0, 1]
```

### 调整subplot周围的间距

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0) # 调整间距
```

### 颜色、标记和线性

```python
# 查看所有线型的合集
plt.plot?
```

```python
from numpy.random import randn
plt.plot(randn(30).cumsum(), 'ko--')
```

```python
# 或者
plt.plot(randn(30).cumsum(), color='k', linestyle='dashed', marker='o') # cumsum()累加函数
```

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k--', drawstyle='steps-post', label='step-post')
plt.legend(loc='best')
```
